## 1. Оптическое распознавание набора данных рукописных цифр

```python
from sklearn.datasets import load_digits
import pandas as pd
import matplotlib.pyplot as plt

digits = load_digits()
```

#### Изучение данных

```python
print(digits.DESCR)
```

```python
data = pd.DataFrame(digits.data)
data.head()
```

#### Разделение данных на обучающий и тестовые наборы

```python
from sklearn.model_selection import train_test_split

x = digits.data
y = digits.target

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

#### Исследование данных с помощью визуализации

```python
fig, axes = plt.subplots(2, 5, figsize=(10, 5))

for i, ax in enumerate(axes.flat):
    ax.matshow(digits.images[i])
    ax.set_title(f'Цифра: {digits.target[i]}')
plt.show()
```

#### Построение модели по методу k-ближайших средних

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
```

#### Прогноз модели

```python
y_pred = knn.predict(x_test)
print("Прогнозы для тестового набора:\n {}".format(y_pred))
```

#### Оценка качества модели

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Правильность на тестовом наборе: {accuracy:.2f}")
print(report)
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 5, figsize=(10, 3))

for i in range(5):
    axes[i].imshow(x_test[i].reshape(8, 8), cmap='gray')
    axes[i].set_title(f"Прогноз: {y_pred[i]}\nИстинное: {y_test[i]}")
    axes[i].axis('off')

plt.suptitle("Прогнозы модели для тестовых изображений")
plt.show()
```

#### Вывод

Модель K-ближайших соседей показала отличные результаты на задаче классификации рукописных цифр.
Модель подходит для задачи распознавания рукописных цифр.

## 2. Набор данных по распознаванию вин

```python
from sklearn.datasets import load_wine
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

wine = load_wine()
```

#### Изучение данных

```python
print(wine.DESCR)
data = pd.DataFrame(wine.data, columns=wine.feature_names)
data.head()
```

#### Разделение данных на обучающий и тестовые наборы

```python
from sklearn.model_selection import train_test_split

x = wine.data
y = wine.target

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

#### Исследование данных с помощью визуализации

```python
data = pd.DataFrame(wine.data, columns=wine.feature_names)
data['target'] = wine.target

data.drop('target', axis=1).hist(bins=15, figsize=(15, 10), edgecolor='black')
plt.suptitle("Распределение признаков набора данных о вине")
plt.show()
```

#### Построение модели по методу k-ближайших средних

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
```

#### Прогноз модели

```python
y_pred = knn.predict(x_test)
print("Прогнозы для тестового набора:\n {}".format(y_pred))
```

#### Оценка качества модели

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Правильность на тестовом наборе: {accuracy:.2f}")
print(report)
```

```python
import matplotlib.pyplot as plt
import numpy as np

index = np.arange(10)
plt.figure(figsize=(10, 6))

plt.plot(index, y_pred[:10], marker='o', linestyle='-', color='r', label='Прогноз')
plt.plot(index, y_test[:10], marker='x', linestyle='--', color='b', label='Истинное значение')

plt.title("Прогнозы модели и истинные значения для первых 10 наблюдений")
plt.xlabel("Наблюдения")
plt.ylabel("Класс")
plt.xticks(index, [f'{i+1}' for i in index])
plt.legend()

plt.show()
```

#### Вывод

Модель k-ближайших соседей показала хорошие результаты для задачи классификации вин. Несмотря на несколько ошибочных предсказаний, большинство прогнозов были верными.

## 3. Набор данных по раку груди висконсин (диагностический)

```python
from sklearn.datasets import load_breast_cancer
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

breast_cancer = load_breast_cancer()
```

#### Изучение данных

```python
print(breast_cancer.DESCR)
data = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
data.head()
```

#### Разделение данных на обучающий и тестовые наборы

```python
from sklearn.model_selection import train_test_split

x = breast_cancer.data
y = breast_cancer.target

x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

#### Исследование данных с помощью визуализации

```python
data = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
data['target'] = breast_cancer.target

data.drop('target', axis=1).hist(bins=15, figsize=(15, 10), edgecolor='black')
plt.suptitle("Распределение признаков набора данных по раку груди")
plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```

#### Построение модели по методу k-ближайших средних

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
```

#### Прогноз модели

```python
y_pred = knn.predict(x_test)
print("Прогнозы для тестового набора:\n {}".format(y_pred))
```

#### Оценка качества модели

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Правильность на тестовом наборе: {accuracy:.2f}")
print(report)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.title("Матрица неточности")
plt.xlabel("Предсказанный класс")
plt.ylabel("Истинный класс")
plt.show()
```

#### Вывод

Модель K-ближайших соседей показала высокую точность в классификации данных о раке груди. Модель KNN достигла точности 92% на тестовых данных. Это показывает, что она способна правильно определить, является ли опухоль злокачественной или доброкачественной.
