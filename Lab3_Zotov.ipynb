```python
# Два графика на диаграмме в индивидуальных масштабах

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

X = np.linspace(0, 4*np.pi, 400)

leg_01 = 'exp(-0.1*t)*sin(2*t)'

Y_01 = np.sin(2*X)*np.exp(-0.1*X)

leg_02 = "10*sin(0.5*t)+cos(10*t)"

Y_02 = 10*np.sin(0.5*X)+np.cos(10*X)

# размеры изображения диаграммы
mpl.rcParams['figure.figsize'] = (8.0, 6.0)

# построение диаграммы

ax_01 = plt.axes()

line_01, = ax_01.plot(X, Y_01, 'b-')

ax_01.axis([-1, 13, -1.1, 1.1])

ax_01.grid(color = 'b', linewidth = 1)

ax_01.set_xlabel('Время t sec', fontsize = 'x-large')
ax_01.set_ylabel(leg_01, color = 'b', fontsize = 'x-large')

ax_01.set_title('Гармонические сигналы' + leg_01 + ' и ' + leg_02)

ax_02 = ax_01.twinx()

line_02, = ax_02.plot(X, Y_02, 'r-')

ax_02.axis([-1, 13, -12, 12])

ax_02.set_ylabel(leg_02, color = 'r', fontsize = 'x-large')

ax_02.legend((line_01, line_02), (leg_01, leg_02), loc = 'best')

ax_02.grid(color = 'r')
plt.show()
```

```python
# Параметрический график
# Построим x=Ax*cos(Wx*t), y=Ay*sin(Wy*t)

import matplotlib.pyplot as plt
import numpy as np

Ax=1; Ay=1; Wx=2; Wy=3

# независимая переменная
t = np.linspace(0,4*np.pi, 100)

# параметр aspect(ratio)
axx = plt.subplot(111)
axx.set_aspect(1)

# изменение шага сетки на графике
plt.xticks([i for i in np.linspace(-1.1*Ax, 1.1*Ax, 5)])
plt.yticks([i for i in np.linspace(-1.1*Ay, 1.1*Ax, 10)])

# заголовок диаграммы
plt.title("Параметрический график " \
          " Ax={0} Wx={1} Ay={2} Wy={3}\n" \
          "          (фигуры Лиссажу)"  \
               .format(Ax, Wx, Ay, Wy))

plt.plot(Ax * np.sin(Wx*t), Ay*np.cos(Wy*t))
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)

y = x**2-3*x+2

plt.figure(figsize=(8, 6), dpi=80)

plt.title('График квадратичной функции', fontsize=16, color='darkblue', pad=20)

plt.plot(x, y, label='y = x^2-3x+2', color='blue', linewidth=2)

plt.xlabel('X', fontsize=12, color='green')
plt.ylabel('Y', fontsize=12, color='green', rotation=0)

ax = plt.gca()
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.grid(True, linestyle=':', color='gray')


plt.legend(loc='upper right')

plt.show()
```

### 3. Построить график произвольной гармонической функции

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 400)

y = np.exp(-0.1 * x) * np.sin(2 * np.pi * x)

plt.figure(figsize=(8, 6), dpi=80)

plt.title('График гармонической функции (синус с затуханием)', fontsize=16, color='darkblue', pad=20)

plt.plot(x, y, label='y = e^(-0.1x)*sin(2πx)', color='blue', linewidth=2)

plt.xlabel('X', fontsize=12, color='green')
plt.ylabel('Y', fontsize=12, color='green', rotation=0)

ax = plt.gca()
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.grid(True, linestyle=':', color='gray')

plt.legend(loc='upper right')

plt.show()
```

### 4. Вывести графики п.2 и 3 в разные поля при помощи функции subplot()

### 6.	Построить график функции, заданной параметрически,
\begin{equation*}
x(t) = \cos(3t), y = \sin(2t), t \in [0, 4\pi]
\end{equation*}

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 4*np.pi, 400)

x = np.cos(3*t)
y = np.sin(2*t)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label=r'$x(t) = \cos(3t), y = \sin(2t)$', color='blueviolet')

plt.title(r'Параметрический график: $x(t) = \cos(3t)$, $y = \sin(2t)$', fontsize=16)
plt.xlabel(r'$x(t)$', fontsize=12)
plt.ylabel(r'$y$', fontsize=12)

plt.grid(True, linestyle=':')
plt.legend()

plt.show()
```

### 7.	Построить столбиковую диаграмму по данным статистического комитета КР по прожиточному минимуму, представленному в таблице

\begin{array}{|c|c|c|c|c|c|c|c|c|c|c|}
\hline
\text{Наименование показателей} & 2012 & 2013 & 2014 & 2015 & 2016 & 2017 & 2018 & 2019 & 2020 & 2021 & 2022 \\
\hline
\text{Кыргызская Республика} & 4341.15 & 4599.21 & 4981.51 & 5182.99 & 4794.34 & 4900.79 & 4792.54 & 4806.32 & 5358.53 & 6268.31 & 7178.32 \\
\hline
\end{array}

```python
import numpy as np
import matplotlib.pyplot as plt

years = np.arange(2012, 2023)
living_wage = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]

plt.figure(figsize=(10, 6))
plt.bar(years, living_wage, color='darkblue')

plt.title('Прожиточный минимум в КР', fontsize=16)
plt.xlabel('Год', fontsize=12)
plt.ylabel('Величина', fontsize=12)

plt.grid(axis='y', linestyle='--')
plt.tight_layout()
plt.show()
```

### 8.	Построить круговой график по данным п. 7

```python
years = np.arange(2012, 2023)
living_wage = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]

plt.figure(figsize=(10, 8), dpi=80)

plt.pie(
    living_wage,
    labels=years,
    autopct=lambda p: '{:.2f}'.format(p * sum(living_wage) / 100),
    startangle=140,
    colors=plt.cm.plasma(np.linspace(0, 1, 10)),
    explode=[0.01] * len(years)
)

plt.title('Прожиточный минимум в КР', fontsize=16)

plt.show()
```
